```python
import csv

# Форматируем csv-файл с записью спектров Калия
with open('/Users/aleksandrbogdanov/MyJupyterNotebook/Datasets/2024 2/k.mdb.csv', 'r') as file:
    csvresd = csv.reader(file)
    format_rows = [] # Здесь будем хранить значения


    for row in csvresd:
        print(type(row[0]))
        if row[0].startswith('SPCH'):
            # Применяем strip к единственному элементу строки
            form_row = row[0].replace("SPCH 131 0 0 1234 spectre_type{5} ", '')
            #form_row = form_row[0].strip('}')
            format_rows.append(form_row)

            
    
    
print(format_rows)
# Можем заметить, что каждая строка файла расценивается как один элемент причем формата str
# Нужно разбить значения и привести в формат float
```

```python
# Меняем тип значений со str на float, создаем массив типа list

K_format_rows = [] # Здесь будут храниться значения

for row in format_rows:
    digit_row = ''
    K_canal = []
    for digit in row:
        digit_row+=digit
        if digit == ' ':
            K_canal.append(float(digit_row))
            digit_row = ''
    
    K_format_rows.append(K_canal)

    
print(K_format_rows)
```

```python
# Переводим list в массив numpy. Суммируем значения спектров 

import numpy as np

K_array_10 = np.array(K_format_rows)
K_array = np.sum(K_array_10, axis=0)
    
print(K_array_10)
print(K_array[:15])
print(K_array.shape)
```

Обработка измеренного спектра

```python
# Проделываем аналогичную процедуру для измеренного спектра непосредственно 

with open('/Users/aleksandrbogdanov/MyJupyterNotebook/Datasets/2024 2/quest.mdb.csv', 'r') as file: 
    csvresd = csv.reader(file)
    format_rows = []

    for row in csvresd:
        if row[0].startswith('SPCH'):
            # Применяем strip к единственному элементу строки
            form_row = row[0].replace("SPCH 131 0 0 1234 spectre_type{5} ", '')
            #form_row = form_row[0].strip('}')
            format_rows.append(form_row)
            
Q_format_rows = []

for row in format_rows:
    digit_row = ''
    Q_canal = []
    for digit in row:
        digit_row+=digit
        if digit == ' ':
            Q_canal.append(float(digit_row))
            digit_row = ''
            
    Q_format_rows.append(Q_canal)

Q_array = np.array(Q_format_rows)

# получаем 52 записи (по 5 см) для 1024 каналов - матрица 52 х 1024 
print(Q_array[:15])
print(Q_array.shape)
```

Визуализация

```python
# Визуализация элементраных спектров и измеренного в лог - масштабе

import matplotlib.pyplot as plt

K_array_log = np.log(K_array)
Th_array_log = np.log(Th_array)
U_array_log = np.log(U_array)

# Здесь в Q_array ставим значения от 0 до 51 - наши измерения 

Q_log = np.log(Q_array[32])

canals = np.arange(1, 1025, 1)

plt.figure(figsize=(15, 10))
plt.plot(canals, K_array_log, c = 'b')
plt.plot(canals, Th_array_log, c = 'r')
plt.plot(canals, U_array_log, c= 'g')
plt.plot(canals, Q_log, c = 'm')

plt.xlabel('Каналы')
plt.ylabel('Спектр')

plt.grid()

```

Определим концентрации МНК

```python
# С помощью МНК найдем концентрации элементов - U, Th, K по формуле Conc = (S.T * S)^-1 * S.T * Q_array
# Используем псевдообратную матрицу Мура-Пенроуза, чтобы исключить возможность Singular matrix,
# когда матрица (S.T * S)^-1 вырожденная - не существует 

Conc = np.linalg.pinv(Spectrum.T) @ Q_array.T

Conc_n = Conc.T
print(Conc_n)
```

```python
# Визуализируем концентрации, чтобы определить, какой элемент преобладал в образце при записи

K_Conc = Conc_n[:,0] # Столбец 0 - К
Th_Conc = Conc_n[:, 1] # Столбец 1 - Th
U_Conc = Conc_n[:, 2] # Столбец 2 - U



samp = np.arange(1, 53, 1) # Массив для 52 записей

plt.figure(figsize=(25,15))

plt.plot(samp, K_Conc, label = 'K', c = 'r')
plt.plot(samp, Th_Conc, label = 'Th', c = 'b')
plt.plot(samp, U_Conc, label = 'U', c = 'g')
plt.xticks(range(1,53,1))

plt.xlabel('Шаг измерения')
plt.ylabel('Концентрация элементов')

plt.legend(fontsize = 20)

plt.grid()
```
