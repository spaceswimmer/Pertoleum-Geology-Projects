```python
# Достанем слова, которые будем использовать в расстоянии Левенштейна
# Одна "буква" - целое описание породы
rocks_taeznaya_2 = taeznaya_2['Rock'] + taeznaya_2['Color'] + ' ' + taeznaya_2['ТЕКСТУРА'] + ' ' + taeznaya_2['Включения органогенные'] + ' ' + taeznaya_2['Включения минеральные'] + ' ' + taeznaya_2['НЕФТЕНАСЫЩЕНИЕ'] + ' ' + taeznaya_2['ПОДОШВА СЛОЯ']
rocks_taeznaya_3 = taeznaya_3['Rock'] + taeznaya_3['Color'] + ' ' + taeznaya_3['ТЕКСТУРА'] + ' ' + taeznaya_3['Включения органогенные'] + ' ' + taeznaya_3['Включения минеральные'] + ' ' + taeznaya_3['НЕФТЕНАСЫЩЕНИЕ'] + ' ' + taeznaya_3['ПОДОШВА СЛОЯ']
rocks_taeznaya_2
```

```python
# Создадим из записей - списки
rocks_2_list = [rock for rock in rocks_taeznaya_2]
rocks_3_list = [rock for rock in rocks_taeznaya_3]
print(rocks_3_list)
print('-----------')
print(rocks_2_list)
```

```python
# Создаем каркас матрицы Левенштейна
levenstein_matrix = np.zeros((len(rocks_3_list)+1, len(rocks_2_list)+1))
levenstein_matrix[0,:] = np.arange(len(rocks_2_list)+1)
levenstein_matrix[:,0] = np.arange(len(rocks_3_list)+1)
levenstein_matrix
```

```python
levenstein_matrix.shape[0]
```

```python
# Заполняем матрицу левенштейна по алгоритму

for i in range(1, levenstein_matrix.shape[0]):
    
    for j in range(1, levenstein_matrix.shape[1]):
        
        left_val = levenstein_matrix[i, j-1] + 1
        upper_val = levenstein_matrix[i-1, j] + 1
        
        if rocks_2_list[j-1] == rocks_3_list[i-1]:
            diag_val = levenstein_matrix[i-1, j-1]
        else:
            diag_val = levenstein_matrix[i-1, j-1] + 1
        
        insert_value = min([left_val, upper_val, diag_val])
        
        levenstein_matrix[i,j] = insert_value
        
        
levenstein_matrix
```

```python
# Посмотрим на хит мэп, тут будет видно минимальное расстояние от последнего элемента до первого - наиболее холодные оттенки

import seaborn as sns
sns.heatmap(levenstein_matrix, annot=True)
```

```python
# Цель данной ячейки кода - изобразить, как мы двигаемся по матрице Левенштейна в обратном порядке
# То есть метод редакционного предписания
# Также соберем информацию о корреляции пластов, понятно, что пласты из Таежная 3 будут выклиниваться, но какие?
# Собрав эту информацию, сможем сопоставить с пластами и понять, какие выклиниваются, а какие коррелируют

import matplotlib.pyplot as plt
# Построение тепловой карты с seaborn
sns.heatmap(levenstein_matrix, cmap='viridis', annot=True) 

# Здесь будем хранить информацию о коррелировании пластов
correlate = [] 

i = levenstein_matrix.shape[0]-1
j = levenstein_matrix.shape[1]-1

search_value = levenstein_matrix[i, j]

# Каждый раз визуализируем наш путь

plt.plot([13,14], [16,17], color='red', linestyle='-', linewidth=2, marker='o', markersize=3, markeredgecolor='red')

while i != 0 and j != 0:
        
        left_val = levenstein_matrix[i, j-1]
        upper_val = levenstein_matrix[i-1, j]
        diag_val = levenstein_matrix[i-1, j-1]
        
        # Когда породы коррелируют (диагонали равны) - ничего не меняем
        
        if (search_value == diag_val and left_val == upper_val) or search_value == upper_val:
            
            # Рисование ломаной линии
            plt.plot([j-1,j], [i-1,i], color='red', linestyle='-', linewidth=2, marker='o', markersize=3, markeredgecolor='red')
            
            correlate.append('Пласты коррелируют')
            
            i -= 1
            j -= 1
            
            search_value = levenstein_matrix[i][j]
         
        # Когда необходимо вставить букву = выкливается пласт Таежная 3
        elif min([left_val, upper_val, diag_val]) == upper_val:
            
            # Рисование ломаной линии
            plt.plot([j,j+1], [i-1,i], color='red', linestyle='-', linewidth=2, marker='o', markersize=3, markeredgecolor='red')
            
            correlate.append('Пласт Таежная_3 - выклинивается')
            
            i -= 1
            
            search_value = levenstein_matrix[i][j]
            
            
        # Когда необходимо поменять буквы - корреляция пластов
        
        elif min([left_val, upper_val, diag_val]) == diag_val:
            
            # Рисование ломаной линии
            plt.plot([j-1,j], [i-1,i], color='red', linestyle='-', linewidth=2, marker='o', markersize=3, markeredgecolor='red')
            
            correlate.append('Пласты коррелируют')

            
            i -= 1
            j-= 1
            
            search_value = levenstein_matrix[i][j]
            
        # Когда необходимо удалить букву - у нас такой вариант точно не встретится в задаче     
        elif min([left_val, upper_val, diag_val]) == left_val:
            
            # Рисование ломаной линии
            plt.plot([j-2,j-1], [i-1,i], color='red', linestyle='-', linewidth=2, marker='o', markersize=3, markeredgecolor='red')
            
            correlate.append('Пласт Таежная_2 - выклинивается')

            
            j -= 1
            
            search_value = levenstein_matrix[i][j]

plt.plot([0,1], [0,1], color='red', linestyle='-', linewidth=2, marker='o', markersize=3, markeredgecolor='red')

plt.title('Редакционное предписание')

# Отображение графика
plt.show()
```

```python
# Визуализируем корреляцию

# массивы с глубинами и лит индексами (2 - таежная 2 | 3 - таежная 3)
dep_2 = taeznaya_2['Depth']
dep_2 = [1*i for i in dep_2]
dep_3 = [1*i for i in taeznaya_3['Depth']]
rock_2 = [i for i in taeznaya_2['Rock']]
rock_3 = [i for i in taeznaya_3['Rock']]

# Литология
plt.step(np.array(dep_2), np.array(taeznaya_2['Rock']), c = 'b', lw = 2, label = 'Taeznaya_2')
plt.step(np.array(dep_3)+10, np.array(taeznaya_3['Rock']) + '10', c = 'm', lw = 2, label = 'Taeznaya_3')

# список коррелирующих (и нет) пластов
path = [[0,0], [1,1], [1, 2], [1, 3], [2, 4], [3,5], [4,6], [5, 7], [6,8], [7,9], [8,10], [8,11], [9,12], [10,13], [11,14], [12,15]]

# Строим корреляции
for p in path:
    if (p[0] == 1 and p[1] == 2) or (p[0] == 1 and p[1] == 3) or (p[0] == 8 and p[1] == 11):
        plt.plot([dep_2[p[0]], dep_3[p[1]]+10], [rock_2[p[0]], rock_3[p[1]]+'10'], color = 'r')
    else:    
        plt.plot([dep_2[p[0]], dep_3[p[1]]+10], [rock_2[p[0]], rock_3[p[1]]+'10'], color = 'g')
        
plt.xlabel('Depth')
plt.ylabel('Lithology', fontsize = 10)
plt.title('Корреляция пластов Таежная_2 - Таежная_3')

plt.legend()
```
